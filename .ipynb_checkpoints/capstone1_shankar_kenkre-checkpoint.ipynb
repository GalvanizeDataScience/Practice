```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np

# Get HTML from the first page

r_first_page = requests.get('https://www.payscale.com/college-salary-report/bachelors')



soup=BeautifulSoup(r_first_page.text,'html.parser')

table = soup.select("table.data-table")[0]
columns = table.find("thead").find_all("th")

#column_names = [c.string for c in columns]
table_rows = table.find("tbody").find_all("tr")
column_names = [c.get_text() for c in table.select('thead th')]

l=[]

for tr in table_rows:
    td = tr.select(".data-table__value")
    row = [str(tr.get_text()).strip() for tr in td]
    l.append(row)
```

```python
# collect data from pag 2 to page 63

for i in range(2,64):
    

    next_page = requests.get(f'https://www.payscale.com/college-salary-report/bachelors/page/{i}')

    soup=BeautifulSoup(next_page.text,'html.parser')

    table = soup.select("table.data-table")[0]
    columns = table.find("thead").find_all("th")

    #column_names = [c.string for c in columns]
    table_rows = table.find("tbody").find_all("tr")
    column_names = [c.get_text() for c in table.select('thead th')]

    for tr in table_rows:
        td = tr.select(".data-table__value")
        row = [str(tr.get_text()).strip() for tr in td]
        l.append(row)

    
df_pay = pd.DataFrame(l,columns=column_names)
```

```python
df_tuition_few_columns=df_tuition[[
'UnitID',
'Institution Name',
'State abbreviation (HD2017)',
'Percent paying sticker price',
'Institution (entity) name (HD2017)',
'Institution name alias (HD2017)',
'City location of institution (HD2017)',
'Longitude location of institution (HD2017)',
'Latitude location of institution (HD2017)',
'Retention full-time total',
'Percent admitted - total (DRVADM2017)',
'Grand total (GR2017  4-year institutions  Completers within 150% of normal time)',
'Grand total (EF2017  All students  Undergraduate total)',
'Total men (EF2017  All students  Undergraduate total)',
'Total women (EF2017  All students  Undergraduate total)',
'American Indian or Alaska Native total (EF2017A  All students  Undergraduate total)',
'Asian total (EF2017A  All students  Undergraduate total)',
'Black or African American total (EF2017A  All students  Undergraduate total)',
'Hispanic total (EF2017A  All students  Undergraduate total)',
'Native Hawaiian or Other Pacific Islander total (EF2017A  All students  Undergraduate total)',
'White total (EF2017A  All students  Undergraduate total)',
'Two or more races total (EF2017A  All students  Undergraduate total)',
'Race/ethnicity unknown total (EF2017A  All students  Undergraduate total)',
'Nonresident alien total (EF2017A  All students  Undergraduate total)',
'Total price for in-state students living on campus 2018-19 (DRVIC2018)',
'Total price for in-state students living off campus (not with family)  2018-19 (DRVIC2018)',
'Total price for in-state students living off campus (not with family)  2019-20 (DRVIC2019)'
]]
```

```python
df_tuition_final = df_tuition_few_columns.dropna(axis='index', how='all')
```

```python
df_pay.rename(columns={'School Name':'school_name'},inplace=True)
```

```python
df_tuition_final.rename(columns={'Institution Name':'school_name',
'Grand total (EF2017  All students  Undergraduate total)':'total_students',
'Total men (EF2017  All students  Undergraduate total)' : 'totalmen',
'Total women (EF2017  All students  Undergraduate total)' : 'totalwomen',
'American Indian or Alaska Native total (EF2017A  All students  Undergraduate total)' : 'totalamreicanindian',
'Asian total (EF2017A  All students  Undergraduate total)' : 'totalasian',
'Black or African American total (EF2017A  All students  Undergraduate total)' : 'totalblack',
'Hispanic total (EF2017A  All students  Undergraduate total)' : 'totalhispanic',
'Native Hawaiian or Other Pacific Islander total (EF2017A  All students  Undergraduate total)' : 'totalhawaiian',
'White total (EF2017A  All students  Undergraduate total)' : 'totalwhite',
'Two or more races total (EF2017A  All students  Undergraduate total)' : 'totalmixedrace',
'Race/ethnicity unknown total (EF2017A  All students  Undergraduate total)' : 'totalunknown',
'Nonresident alien total (EF2017A  All students  Undergraduate total)' :'totalalien'},inplace=True)
```

```python
df_work = pd.merge(df_pay, df_tuition_final, on = 'school_name', how='inner')
```

```python
df_work.columns
```

```python
df_work['%men'] = df_work.apply(lambda r : r.totalmen * 100 / r.total_students, axis=1).round(2)
df_work['%women'] = df_work.apply(lambda r : r.totalwomen * 100 / r.total_students, axis=1).round(2)
df_work['%amreicanindian'] = df_work.apply(lambda r : r.totalamreicanindian * 100 / r.total_students, axis=1).round(2)
df_work['%asian'] = df_work.apply(lambda r : r.totalasian * 100 / r.total_students, axis=1).round(2)
df_work['%black'] = df_work.apply(lambda r : r.totalblack * 100 / r.total_students, axis=1).round(2)
df_work['%hispanic'] = df_work.apply(lambda r : r.totalhispanic * 100 / r.total_students, axis=1).round(2)
df_work['%hawaiian'] = df_work.apply(lambda r : r.totalhawaiian * 100 / r.total_students, axis=1).round(2)
df_work['%white'] = df_work.apply(lambda r : r.totalwhite * 100 / r.total_students, axis=1).round(2)
df_work['%mixedrace'] = df_work.apply(lambda r : r.totalmixedrace * 100 / r.total_students, axis=1).round(2)
df_work['%unknown'] = df_work.apply(lambda r : r.totalunknown * 100 / r.total_students, axis=1).round(2)
df_work['%alien'] = df_work.apply(lambda r : r.totalalien * 100 / r.total_students, axis=1).round(2)
```
